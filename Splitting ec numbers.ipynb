```python
DATAPATH = '/Users/raouldias/Desktop/Masters Project/Datasets/new_taxa.csv'
dataframe = pd.read_csv(DATAPATH, delimiter=',', header=None, skiprows=1, names= ['ec_num'
                                                                                  , 'organism'
                                                                                  , 'km_value'
                                                                                  ,'substrate'
                                                                                  , 'comments'
                                                                                  , 'rando'
                                                                                  , 'tax_id'
                                                                                  , 'superkingdom'
                                                                                  , 'phylum'
                                                                                  , 'class'
                                                                                  , 'order'
                                                                                  , 'family'
                                                                                  , 'genus'
                                                                                  , 'species'
                                                                                  ])
```
